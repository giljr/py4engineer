# 01 - Given the function z to respect to x, determine the extreme points:

image.png







### Let's get down to work! Take the Diferentiation of z to the respect of x

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
```

```python
x= sy.symbols('x')
Z = x**3 - 2*x**2 - x + 2
Zx = sy.diff(Z, x)
print(f'Zx = {Zx}')
```

###Equations


 $ Z= x^3  2x^2  x + 2 $ 

 $ Zx = 3x^2 - 4x - 1 = 0 $

# 02 - Given the function z to respect to x, determine the extreme points:
### $ z = x^3 - \frac{27}{2}x^2 $

image.png

### Let's get down to work! Take the Diferentiation of z to the respect of x

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
```

```python
x= sy.symbols('x')
Z = x**3 - (27/2)*(x**2)
Zx = sy.diff(Z, x)
print(f'Zx = {Zx}')
```

###Equations


 $ Z= x^3  \frac{27}{2}x^2 $ 

 $ Zx = 3x^2 - 27x $

### Roots of a Function

```python
from sympy import solve
solve(3*x**2 - 27.0*x)
```

##Interpretation - Roots:
$ If  \hspace{.5cm} x = 0 \hspace{.5cm} \mapsto \hspace{.5cm} f(0) = 0  \hspace{.5cm} \mapsto  \hspace{.5cm}P(0,0) $

$ If  \hspace{.5cm} x = 9 \hspace{.5cm} \mapsto \hspace{.5cm} f(9) = 9^3 - \frac{27}{2} * 9^2 = -\frac{729}{2}\hspace{.5cm} \mapsto  \hspace{.5cm}P(9,-\frac{729}{2}) $


##To classify the critical points, we need to calculate the second order derivative of the function:

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
```

```python
x, y = sy.symbols('x y')
Z = Z = x**3 - (27/2)*(x**2)
Zx = sy.diff(Z, x)
Zxx = sy.diff(Zx, x)
print(f'Zx = {Zx}\nZxx = {Zxx}')
```

##Interpretation - Extreme Points:
$ For  \hspace{.5cm}P(0,0) \mapsto Zxx = 6*0-27 \mapsto Zxx < 0\hspace{.5cm} Them\hspace{.5cm} MAX$

$ For  \hspace{.5cm}P(9,-364.5) \mapsto Zxx = 6*9-27 \mapsto Zxx > 0\hspace{.5cm} Them\hspace{.5cm} MIN$
