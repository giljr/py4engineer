#3D System - Extreme Points $ z = f(x,y)$  

## 01 - Given the function z to respect to x and y, determine the extreme points:
$ z = 3x^2 + 2xy + y^2 + 10x + 2y +1 $

image.png

### Let's get down to work! Take the Diferentiation of z to the respect of x and them y

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
```

```python
x, y = sy.symbols('x y')
Z = 3*x**2 + 2*x*y + y**2 + 10*x + 2*y + 1
Zx = sy.diff(Z, x)
Zy = sy.diff(Z, y)
print(f'Zx = {Zx}\nZy = {Zy}')
```

###Linear System


 $ Zx = 6x + 2y + 10 = 0   \mapsto  6x + 2y = -10 $ 

 $ Zy = 2x + 2y + 2 = 0 \mapsto 2x + 2y = -2 $

  

###Solving the Linear Equations

```python
A = np.array([[6, 2], [2, 2]])
b = np.array([-10, -2])
x = np.linalg.solve(A,b)

print(x)
```

###So the points are $x = -2;\hspace{.5cm} y = 1$

###Here are The 3 Equations (y1, y2 and y3):



```python
x, y = sy.symbols('x y')
y1 = 3*x**2 + 2*x*y + y**2 + 10*x + 2*y + 1
y2 = 6*x + 2*y + 10
y3 = 2*x + 2*y + 2
```

###Calculation of The First and the Second Partial Derivatives:

###Discriminant - Formula

$D = Zxx*Zyy - (Zxy)^2 $

It involves the second pure partial derivatives in product, subtracting the second mixed derivative squared

Zxx

```python
#differentiate y1 with respect to x
Zx = sy.diff(y1, x)
#returns resulting expression in factored form
#this matters to our specific problem and looking
#for some kind of structure in the expression
sy.factor(Zx)
```

```python
Zxx = sy.diff(Zx, x)
print(f'Zxx={sy.factor(Zxx)}')
```

Zyy

```python
#differentiate y1 with respect to y
Zy = sy.diff(y1, y)
#returns resulting expression in factored form
#this matters to our specific problem and looking
#for some kind of structure in the expression
sy.factor(Zy)
```

```python
Zyy = sy.diff(Zy, y)
print(f'Zyy={sy.factor(Zyy)}')
```

Zxy

```python
#differentiate y1 with respect to x
Zx = sy.diff(y1, x)
#returns resulting expression in factored form
#this matters to our specific problem and looking
#for some kind of structure in the expression
sy.factor(Zx)
```

```python
Zxy = sy.diff(Zx, y)
print(f'Zxy={sy.factor(Zxy)}')
```

###Determinant - Formula

$D = Zxx*Zyy - (Zxy)^2 $

$D = 6*2 - (2)^2  = 8$


It involves the second pure partial derivatives in product, subtracting the second mixed derivative squared

```python
D = Zxx*Zyy - (Zxy)**2
print(f'D={D}')
```

#### As $ D > 0 \hspace{.5cm} and \hspace{.5cm}Zxx > 0$

##Interpretation - Discriminant:
$ If  \hspace{.5cm} D > 0 \hspace{.5cm} and \hspace{.5cm} Zxx < 0 \mapsto  \hspace{.5cm} MAX $

$ If  \hspace{.5cm} D > 0 \hspace{.5cm} and \hspace{.5cm} Zxx > 0 \mapsto  \hspace{.5cm} MIN$

$ If  \hspace{.5cm} D < 0 \hspace{.5cm}\mapsto  \hspace{.5cm} SADDLE $

$ If  \hspace{.5cm} D = 0 \hspace{.5cm}  \mapsto  \hspace{.5cm} 
NOT-CONCLUSIVE$

###The points (-2,1) represent the Minimum

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
