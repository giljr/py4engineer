### String Interpolation - > Injects a variable into your string printing
#### 2 methods:
             . format()
             . f-strings

### format()

```python
print('This is a string {}'.format('INSERTED'))
```

```python
print('{} {} {}.'.format('brown', 'quick', 'fox') )
```

```python
print('{1} {0} {2}.'.format('brown', 'quick', 'fox') )
```

```python
print('{q} {b} {f}'. format(b = 'brown', q = 'quick', f = 'fox'))
```

### Float formating follows "{value: width.procision f}"

```python
result = 100/777
```

```python
print("The result was {}".format(result))
```

```python
print("The result was {r:1.3f}".format(r = result))
```

```python
print("The result was {r:10.3f}".format(r = result))
```

### f-strings

```python
name = 'j3'
print('His name is {}'.format(name))
```

```python
print(f'His name is {name}')
```

```python
print("That's is all for now, folks o/ {}".format("Thank you!"))
```
