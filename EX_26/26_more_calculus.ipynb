# 01 - Double Integrate:

#$   f(x,y)=\int_{0}^{1}\int_{0}^{1}\ xy^2\ dydx  $

```python
from scipy.integrate import dblquad
import numpy as np

def integrand(y, x):
    'y must be the first argument, and x the second.'
    return  x*y**2

ans, err = dblquad(integrand, 0, 1,
                   lambda x: 0,
                   lambda x: 1)
print( ans)
```

# 02 - Double Integrate:

#$   f(x,y)=\int_{0}^{1}\int_{x^2}^{x}\ 2x + 2y\ dydx  $

```python
from scipy.integrate import dblquad
import numpy as np

def integrand(y, x):
    'y must be the first argument, and x the second.'
    return  2*x+2*y

ans, err = dblquad(integrand, 0, 1,
                   lambda x: x**2,
                   lambda x: x)
print( ans)
```

# 03 - Double Integrate:

#$   f(x,y)=\int_{0}^{1}\int_{-\sqrt{1-y^2}}^{\sqrt{1-y^2}}\ y\ dydx  $

```python
from scipy.integrate import dblquad
import numpy as np
import sympy as sy

def integrand(x, y):
    'y must be the first argument, and x the second.'
    return  y

ans, err = dblquad(integrand, 0, 1,
                   lambda y: -sy.sqrt(1-y**2),
                   lambda y: sy.sqrt(1-y**2))
print( ans)
```

# 04 - Triple Integrate:

#$   f(x,y, z)=\int_{0}^{2}\int_{1}^{x}\int_{0}^{x+y}\ 3xyz\ dzdydx  $

```python
from scipy import integrate

f = lambda z, y, x: 3*x*y*z
integrate.tplquad(f, 0, 2, lambda x: 1, lambda x: x,
                  lambda x, y: 0, lambda x, y: x+y)
```

#06 - Differentiation By Chain Rule

Use Chain Rule to Differentiate this function by hand:)

## $ z(t) = 3x^2 + 7xy + y^2 $
Where:
## $ x = 2t $
And 
## $ y = t^2 $



###Solution:


Chain $ \begin{equation}
  Rule =  \cfrac{z}{ 
       \cfrac{ x \, y}{
       t \,t } }
\end{equation} $

Formula of partials:

$\frac {dz}{dt} =  \frac{\partial z}{\partial x}. \frac{\partial z}{\partial y} + \frac{\partial z}{\partial y}. \frac{\partial y}{\partial t} $

Calculations of partials:

$ \frac{\partial z}{\partial x} = 6x + 7y $

$ \frac{\partial x}{\partial t} = 2 $

$ \frac{\partial z}{\partial y} = 7x + 2y $

$ \frac{\partial y}{\partial t} = 2t $

replacing:

$ (6x+7y)(2) + (7x+2y)(2t) $

$ 12x + 14y + 14 xt + 4t$

$ 12x(2t) + 14(t^2) + 14(2t)t + 4(t^2)t $

$ 24t + 14^2 + 28t^2 + 4t^3 $

 $ 4t^3 + 42t^2 + 24t $




```python
# Working w/ Python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy

x, t = sy.symbols('x t')
Z = 3*(2*t)**2 + 7*2*t*t**2 + (t**2)**2
Zx = sy.diff(Z, t)
Zy = sy.diff(Z, t)
print(f'Zx = {Zx}')
```

#07 - Differentiation By Chain Rule

Use Chain Rule to Differentiate this function by hand:)

## $ z (s) = 3x^2 + 2x $
Where:
## $ x = t + 2s $
And 
## $ y = 3t +  s $


###Solution:


Chain $ \begin{equation}
  Rule =  \cfrac{z}{ 
       \cfrac{ x \, y}{
       st \,st } }
\end{equation} $

Formula of partials:

$\frac {dz}{ds} =  \frac{\partial z}{\partial x}. \frac{\partial x}{\partial s} + \frac{\partial z}{\partial y}. \frac{\partial y}{\partial s} $

Calculations of partials:

$ \frac{\partial z}{\partial x} = 2 $

$ \frac{\partial x}{\partial s} = 2 $

$ \frac{\partial z}{\partial y} = 6y $

$ \frac{\partial y}{\partial s} = 1 $

replacing:

$ (2)(2) + (6y)(1) $

$ 4 + 6(3t + s)$

$ 4 + 18t + 6s $

$  6s + 18t + 4 $




```python
# Or or presenting in a pleasant way:)
from sympy import *
z, t, s=symbols("z t s")
init_printing()
z=3*(3*t+s)**2 + 2*(t+2*s)
diff(z,s)
```

# Gradient - Curl (from https://docs.sympy.org/latest/modules/vector/fields.html)

A curl is a mathematical operator that describes an infinitesimal rotation of a vector in 3D space. 

The direction is determined by the right-hand rule (along the axis of rotation), and the magnitude is given by the magnitude of rotation.

In the 3D Cartesian system, the curl (or Gradient of scalar field) of a 3D vector F , denoted by ∇×F is given by:

$ \nabla f = grad f = \frac{\partial f}{\partial x} \overrightarrow{i} + \frac{\partial f}{\partial y} \overrightarrow{j} + \frac{\partial f}{\partial z} \overrightarrow{k} $

The Del, or ‘Nabla’ operator - written as ∇ is commonly known as the vector differential operator. Depending on its usage in a mathematical expression, it may denote the gradient of a scalar field, the divergence of a vector field, or the curl of a vector field.

Since:
##$f (x, y, z) =x^2 - y +z^2 $, 

Calculate the directional derivative at point: 

##P (1; 2; 1) 

In the direction and direction of the vector:

##$ \overrightarrow{u}  = 4 \overrightarrow{i} - 2 \ \overrightarrow{j} + 4 \overrightarrow{k}  $

###Gradient is a vector that indicates the MAXIMUM GROWTH direction of a scalar function f (x, y, z)


$ \nabla f = grad f = \frac{\partial f}{\partial x} \overrightarrow{i} + \frac{\partial f}{\partial y} \overrightarrow{j} + \frac{\partial f}{\partial z} \overrightarrow{k} $

```python
from sympy import *
F,x,y,z =symbols("F x y z")
init_printing()
F=x**2-y-z**2
print(f'Fx={diff(F,x)}\nFy={diff(F,y)}\nFz={diff(F,z)}')
```

$ \nabla = grad f = \overrightarrow{u} =  2x \overrightarrow{i} - 1 \ \overrightarrow{j} + 2z \overrightarrow{k}  $

The **Directional Derivative** in the direction of $\overrightarrow{u}$  is:

$ \frac{df}{ds} = \nabla f . \overrightarrow{u} $ 

Calculate Unity Vector:

$ \overrightarrow{u} =  \frac{\overrightarrow v}{||\overrightarrow v||} = \frac{4\overrightarrow{i}-2\overrightarrow{j} + 4\overrightarrow{k}}{\sqrt(4^2+(-2)^2+4^2)} = \frac{4\overrightarrow{i}-2\overrightarrow{j} + 4\overrightarrow{k}}{\sqrt(36)} = \frac{2\overrightarrow{i}-1\overrightarrow{j} + 2\overrightarrow{k}}{3}  $

Now Calculate the scalar product:
Given P(1;2;1)

$ \frac{df}{ds} = \nabla f . \overrightarrow{u}   =  (2-\overrightarrow{j} + 2 \overrightarrow{k}) (\frac{2\overrightarrow{i}-\overrightarrow{j} + 2\overrightarrow{k}}{3}) = \frac{2.2-1.(1)+2.2}{3} =  \frac{4+1+4}{3} = \frac{9}{3} = 3  $

$ \frac{df}{ds} = \nabla f . \overrightarrow{u}  = 3$ 

###What does it mean?

It means that the function $ f (x, y, z) =x^2 - y +z^2 $ increases the rate of 3 units of distance when we move away from $P(1;2;1)$ in the given direction: 
$ 4\overrightarrow{i}-2\overrightarrow{j} + 4\overrightarrow{k} $
