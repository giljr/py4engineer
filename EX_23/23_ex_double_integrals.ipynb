## 01 - Integrate:

$   f(x,y)=\int_{0}^{2}\int_{0}^{x^2}\ y\ dydx  $


```python
from scipy.integrate import dblquad
import numpy as np

def integrand(y, x):
    'y must be the first argument, and x the second.'
    return y

ans, err = dblquad(integrand, 0, 2,
                   lambda x: 0,
                   lambda x: x**2)
print( ans)
```

 ## 02 - Integrate:

$   f(x,y)=\int_{0}^{1}\int_{0}^{x^2}\ (x+2)\ dydx  $

```python
from scipy.integrate import dblquad
import numpy as np

def integrand(y, x):
    'y must be the first argument, and x the second.'
    return  x+2

ans, err = dblquad(integrand, 0, 1,
                   lambda x: 0,
                   lambda x: 2)
print( ans)
```

 ## 04 - Integrate:

$   f(x,y)=\int_{0}^{2}\int_{x}^{6-x^2}\ (3x+2y)\ dydx  $

```python
from scipy.integrate import dblquad
import numpy as np

def integrand(y, x):
    'y must be the first argument, and x the second.'
    return  (3*x+2*y)

ans, err = dblquad(integrand, 0, 2,
                   lambda x: x,
                   lambda x: 6-x**2)
print( ans)
```

 ## 05 - Determine the area in the First quadrant of the region bounded by :

###Given parable: $ f(x)=x^3 $ 

###And Line: $ g(x)=4x $ 

$ over $ 

$ (0<=x<=2) $


```python
x = np.linspace(0, 3, 1000)
def f(x): return x**3
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(x, f(x))
plt.axhline(color = 'black')
plt.fill_between(x, f(x), where = [(x > 0) and (x < 2) for x in x], color = 'green', alpha = 0.3)
```

```python
import sympy as sy
x = sy.Symbol('x')
```
